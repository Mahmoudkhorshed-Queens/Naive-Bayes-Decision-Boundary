# Task 1

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model, preprocessing
from sklearn.naive_bayes import GaussianNB as GNB
from sklearn.preprocessing import StandardScaler
```

```python
# TODO
names=['feature_1','feature_2','output']  #name the column
df = pd.read_csv('binclass.txt',delimiter=',',names=names)   #load file binclass 
df.head()   #print 5 first Rows
```

```python
df.shape
```

```python
y = df['output'].values;   #extract 3 rd column from dataframe as labels
name=df.columns[:2]        #extract 1 th and 2 nd column from dataframe as features
X=pd.DataFrame(df,columns=name)   #create a new dataframe for features
X = X.values
```

```python

from matplotlib.colors import ListedColormap       
plt.scatter(X[:,0:1], X[:,1:2], c=y,cmap=ListedColormap(['#FF0000', '#0000FF']),label='Negative')
plt.title("Red Class positive- Blue Class negative ")
plt.xlabel("X1-values")
plt.ylabel("X2-values")
plt.legend()                                                                    #plot scatter
                                                                                # between feature 1 and 2 with 
                                                                                # labels 1 red 0 Blue
plt.show
```

```python
import seaborn as sns
sns.scatterplot(data=df,x='feature_1',y="feature_2", hue=y);
```

```python
Model=GNB()    #train model with Gaussian naive bayes
X_train=X
y_train=df['output']
Model.fit(X_train,y_train)
```

```python
h = .02  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

from matplotlib.colors import ListedColormap
plt.scatter(X_train[:,0:1], X_train[:,1:2], c=y_train,cmap=ListedColormap(['#FF0000', '#0000FF']),label='Negative')
plt.title("Red Class positive- Blue Class negative ")
plt.xlabel("X1-values")
plt.ylabel("X2-values")
plt.legend()
plt.show
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
Z = Model.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0],cmap=plt.cm.Paired,linewidths=(6,))
plt.grid()
```
